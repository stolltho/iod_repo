### 6.3.1. Descriptive statistics

```python
def descriptive_statistics(data, text):
    """
    Function performing descriptive statistics and 
    returning a message on data skewness and kurtosis
    
    Input:
    Array of sample data, data title  
    """
             
    nobs, minmax, mean, variance, skewness, kurtosis = stats.describe(data)
    skew = np.round(skewness, 3)
    kurt = np.round(kurtosis, 3)
    
    print(f'\033[1m{text}:\033[0m')
    print("nobs =", nobs)
    print("minmax =", np.round(minmax, 3))
    print("mean =", np.round(mean, 3))
    print("variance =", np.round(variance, 3))
        
    if skew > 1 or skew < -1:
        print(f'skewness = {skew}: Highly skewed distribution.')
    elif 0.5 <= skew <= 1 or -1 <= skew <= -0.5:
        print(f'skewness = {skew}: Moderately skewed distribution.')
    elif -0.5 < skew < 0.5:
        print(f'skewness = {skew}: Fairly symmetrical distribution.')
    else:
        print(f'skewness = {skew}: Out of range.')
    
    if kurt > 1:
        print(f'kurtosis = {kurt}: Distribution is leptokurtic (too peaked).')
    elif kurt < -1:
        print(f'kurtosis = {kurt}: Distribution is platykurtic (too flat).')
    else:
        print(f'kurtosis = {kurt}: Distribution is mesokurtic (normal).')
```

### 6.3.3. Empirical cumulative distribution function (ECDF)

Compare the ECDF of the data to the theoretical CDF of the Normal distribution

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
# Define data in dict to be checked for normal distribution 
data = {'Male customer': male,
        'Female customer': female}
```

```python
data['Male customer'][:3]
```

```python
# Compute and plot CDF of experimental and simulated data (into separate figures)

for key, value in data.items():
    
    # Access respective dict data
    col = data[key]
    
    # Compute ECDF of experimental data
    x, y = ecdf(col)

    # Compute CDF of simulated data
    np.random.seed(42)
    mean = np.mean(col)
    std = np.std(col)
    samples = np.random.normal(mean, std, size=10000)
    x_theor, y_theor = ecdf(samples)
    
    # Plot ECDF and theoretical CDF
    plt.figure(figsize=(5,4))
    plt.plot(x_theor, y_theor)
    plt.plot(x, y, marker = '.', linestyle = 'none')
    plt.xlabel(f"Profit ({key})", labelpad=10)
    plt.ylabel('CDF')
    plt.legend(('simulated', 'experimental'), loc = 'lower right');
```

**Documentation:**
* ECDF plotting indicates that both datasets are not normally distributed

### 6.3.5. Shapiro-Wilk Test

The Shapiro-Wilk test tests the null hypothesis that the data was drawn from a normal distribution

```python
# Confidence level
alpha = 0.05
```

```python
def test_normality(data, alpha=0.05):
    """
    Function performing Shapiro_Wilk Test and returning 
    a significance message on data normality
    
    Inputs:
    Array of sample data, alpha: confidence level    
    """
    
    # Null Hypothesis Dictionary
    null_hypothesis = {'H0': "Data is normally distributed.",
                       'H1': "Data does not come from a normal distribution."}
    
    statistic, p_value = stats.shapiro(data)
    print("test_statistic =", np.round(statistic, 3))
    print("p_value =", np.round(p_value, 3))
    
    if p_value < alpha:
        print('We reject the null hypothesis.')
        print(null_hypothesis['H1'])
    else:
        print('We accept the null hypothesis.') 
        print(null_hypothesis['H0'])
```

```python
# Test data for normality
test_normality(male)
```

```python
# Test data for normality (here: loop over data in dict)
for key, value in data.items():
    
    # Access respective dict data
    col = data[key]
    
    # Perform test and print result
    print(f'Dataset \033[1m{key}:\033[0m')
    test_normality(col)
    print(f'\n')   
```

**Documentation:**
* All five tests indicate that both datasets are not normally distributed

## 6.5. Student's *t*-test (two-tailed)

```python
# Confidence level
alpha = 0.05
```

```python
def ttest(data1, data2, alpha=0.05):
    """
    Function performing Student's t-test (two-tailed) and returning 
    a significance message
    
    Inputs:
    Two arrays of sample data, alpha: confidence level    
    """
    
    # Null Hypothesis Dictionary
    null_hypothesis = {'H0': "Stores in California made equal profit of selling road bikes to male and female customers.",
                       'H1': "Stores in California did \033[1mnot\033[0m make equal profit of selling road bikes to male and female customers."}
    
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    print("t_statistic =", t_statistic)
    print("p_value =", p_value)
    
    if p_value < alpha:
        print('We reject the null hypothesis.')
        print(null_hypothesis['H1'])
    else:
        print('We accept the null hypothesis.') 
        print(null_hypothesis['H0'])
```

```python
# Use log-transformed data for t-test
ttest(male_log, female_log)
```

## 6.6. Mann-Whitney U-test

The Mann-Whitney U-test is a non-parametric alternative to an independent samples t-test

```python
# Confidence level
alpha = 0.05
```

```python
def mannwhitneyu(data1, data2, alpha=0.05):
    """
    Function performing Mann-Whitney U-test for non-parametric data
    and returning a significance message
    
    Inputs:
    Two arrays of sample data, alpha: confidence level    
    """
    
    # Null Hypothesis Dictionary
    null_hypothesis = {'H0': "Stores in California made equal profit of selling road bikes to male and female customers.",
                       'H1': "Stores in California did \033[1mnot\033[0m make equal profit of selling road bikes to male and female customers."}
    
    U_statistic, p_value = stats.mannwhitneyu(data1, data2)
    print("U_statistic =", U_statistic)
    print("p_value =", p_value)
    
    if p_value < alpha:
        print('We reject the null hypothesis.')
        print(null_hypothesis['H1'])
    else:
        print('We accept the null hypothesis.') 
        print(null_hypothesis['H0'])
```

```python
def ttest_return(data1, data2, alpha=0.05):
    """
    Function performing Student's t-test (two-tailed) and returning 
    a significance message
    
    Inputs:
    Two arrays of sample data, alpha: confidence level    
    """
       
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    
    if p_value < alpha:
        x = 'H$_0$ rejected'
        return x
    else:
        y = 'H$_0$ accepted'
        return y
```
