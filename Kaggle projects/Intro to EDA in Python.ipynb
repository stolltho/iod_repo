### 1. Importing libs

```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

### 2. Loading data

```python
df = pd.read_csv('car_dataset.csv')
df.head()
```

### 4. Dropping irrelevant columns

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
```

```python
df.head(5)
```

### 5. Renaming columns

```python
df = df.rename(columns={"Engine HP": "HP","Engine Cylinders": "Cylinders", "Transmission Type": "Transmission",
                        "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price"})
```

```python
df.head(5)
```

```python
df.columns = map(str.lower, df.columns)
#or: df.columns = [x.lower() for x in df.columns]
```

```python
df.head(5)
```

### 6. Dropping the duplicate rows

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()] #Return boolean Series denoting duplicate rows, by default use all of the columns
print(f'Number of duplicate rows: {duplicate_rows_df.shape}')
```

```python
duplicate_rows_df.head(5)
```

```python
df.count() #count entries
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

### 7. Dropping the missing or null values

```python
df.isnull().sum()
```

```python
df = df.dropna() #okay to drop if only a small number
df.isnull().sum()
```
