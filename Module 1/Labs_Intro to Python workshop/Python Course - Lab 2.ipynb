```python
# Create two arrays of 5 arbitrary numbers 
np.random.seed(12)
height_1 = np.random.randint(1, 10, 5)
height_2 = np.random.randint(1, 10, 5)
# Plot two bar charts using plt.bar and the arrays you've created above
# Label them "Example one" and "example two"
plt.bar([1,3,5,7,9], height_1, label="Example one")
plt.bar([2,4,6,8,10], height_2, label="Example two", color='g')
plt.legend()
# Label x axis "Bar number"
plt.xlabel('Bar number')
# Label y axis "Bar height"
plt.ylabel('Bar height')
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code 
# Create an array of a random normal distribution
np.random.seed(40)
dist = np.random.laplace(loc=15, scale=3, size=100) #loc: mean
"""
The Laplace distribution is similar to the Gaussian/normal distribution, but is sharper at the peak and has fatter tails. 
"""
# Create a histrogram using plt hist()
n, bins, patches = plt.hist(x=dist, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
# Set grid for x and y
# set transperancy of the grid using alpha=0.5
plt.grid(axis='x', alpha=0.5)
plt.grid(axis='y', alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```

```python
# Bonus: Create a scatter plot
np.random.seed(40)
x = np.random.randn(1000) + 5
y = np.random.randn(1000) + 5
plt.scatter(x, y, c='red', s=10, alpha=0.5)
plt.grid(True)
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Lots of dots')
plt.show()
```
