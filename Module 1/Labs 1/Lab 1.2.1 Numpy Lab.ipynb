# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# get matplot to show charts inline
%matplotlib inline
````

````python
np.random.seed(42)
height= np.round(np.random.normal(1.75, 0.20, 7000), 2) #mean, std, size
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight)) #double brackets put a, b in a tuple because column_stack takes only 1 arg
````

````python
print('height:\n',height) 
print('weight:\n',weight)
print('population:\n',population)
````

````python
type(population)
````

````python
population.dtype
````

**8. Create a histogram of height and weight**

````python
help(plt.hist)
````

````python
plt.style.use('ggplot')

fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, sharey=True)

fig.suptitle('Height and weight distribution across population')

ax1.hist(height, bins=50, color='#FFC20A') #coloring for colorblindness
ax1.set_xlabel('height (m)')
ax1.set_ylabel('count')
ax1.text(1.4, 440, r'$\mu=1.75,\ \sigma=0.2$', fontsize=10, color='black')

ax2.hist(weight, bins=50, color='#0C7BDC')
ax2.set_xlabel('weight (kg)')
ax2.text(25, 440, r'$\mu=60.32,\ \sigma=15$', fontsize=10, color='black')

plt.show()

plt.rcParams["figure.figsize"] #check figure size
#plt.rcParams["figure.figsize"] = (20, 10) #set figure size
````

**9. Create a boxplot of height and weight**

````python
help(plt.boxplot)
````

````python
plt.style.use('dark_background')

fig, axs = plt.subplots(ncols=1, nrows=2)
ax1, ax2 = axs

fig.tight_layout(pad=2.5)

c = 'red'
ax1.boxplot(height, vert=False, labels=" ", patch_artist=True, widths=0.3, notch=False,
            boxprops=dict(facecolor='white', color=c, linewidth=2),
            capprops=dict(color=c, linewidth=2),
            whiskerprops=dict(color=c, linewidth=2),
            flierprops=dict(color=c, markeredgecolor='orange'),
            medianprops=dict(color=c, linewidth=2))
ax1.set_xlabel('height (m)', fontsize=15)
ax1.set_title('Height and weight distribution across population', pad=20, fontsize=15)

c = 'green'
ax2.boxplot(weight, vert=False, labels=" ", patch_artist=True, widths=0.3, notch=False,
            boxprops=dict(facecolor='white', color=c, linewidth=2),
            capprops=dict(color=c, linewidth=2),
            whiskerprops=dict(color=c, linewidth=2),
            flierprops=dict(color=c, markeredgecolor='lightgreen'),
            medianprops=dict(color=c, linewidth=2))
ax2.set_xlabel('weight (kg)', fontsize=15)
        
plt.show()
````

**10. Create scatterplot of height and weight**
