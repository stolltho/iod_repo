# Statistics - Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# plot frequency of some observations using pie chart
labels = 'Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts'
frequency = [15, 32, 12, 7]

fig1, ax1 = plt.subplots()
explode = (0, 0, 0.2, 0)  # only "explode" the 3rd slice (i.e. 'Data Scientist')
ax1.pie(frequency, 
        explode=explode,
        labels=labels,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.rcParams["figure.figsize"] = (4,3)

plt.show()
```

```python
help(plt.bar)
```

```python
# plot frequency of some observations using line chart
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
 
# Create a bar plot
plt.xticks(y_pos, objects)
plt.plot(performance)
plt.ylabel('Usage')
plt.title('Programming language usage')
plt.rcParams["figure.figsize"] = (4,3)
 
plt.show()
```

```python
# plot frequency of some observations using bar chart
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
 
# Create a bar plot
plt.xticks(y_pos, objects)
plt.bar(objects, performance) 
plt.ylabel('Usage')
plt.title('Programming language usage')
plt.rcParams["figure.figsize"] = (4,3)

plt.show()
```
